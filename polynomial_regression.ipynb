## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')

## Don't include the first column since the Level column encodes the data already
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
print(y)
```

## Training the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

## Training the Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures

# We want to have a matrix of features/columns that will have X1, X1^2, ... X1^n
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)

# Finally build a linear regression model with the polynomial features
poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_poly, y)
```

## Visualizing the Linear Regression Results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Position Salaries (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary ($)')
plt.show()
```

## Visualizing the Polynomial Regression Results

```python
plt.scatter(X, y, color='red')

# Because the linear regressor was fitted with the multi-polynomial-features, it must match that shape
plt.plot(X, poly_lin_reg.predict(X_poly), color='green')
plt.title('Position Salaries (Polynomial Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary ($)')
plt.show()
```

## Predicting Results with Linear Regression

```python
# E.g. let x = 6.5 to see if $150,000 is reasonable
lin_reg.predict([[6.5]])
```

## Predicting Results with Polynomial Regression

```python
poly_lin_reg.predict(poly_reg.fit_transform([[6.5]]))
```
